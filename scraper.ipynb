```python
import json

def scrape(film_lst):
    json_lst = []
    for film in film_lst:
        filmlink = 'https://letterboxd.com/film/' + film + '/'
        thisfilmRep = requests.get(filmlink)
        thisfilmsoup = bs4.BeautifulSoup(thisfilmRep.text, 'html.parser')
        thisfilminfo = thisfilmsoup.find_all(type='application/ld+json')[0].string
        index1 = thisfilminfo.find('{')
        index2 = thisfilminfo.find('/*', index1+1)
        thisfilminfo = thisfilminfo[index1:index2]
        filmjson = json.loads(thisfilminfo)
        try: genre = filmjson['genre']
        except: genre = ['Unknown']
        json_lst.append(filmjson)
    return json_lst
    

    
```

```python
def scrape_themes(film_lst):
    films_themes = []
    for film in film_lst:
        filmlink = 'https://letterboxd.com/film/' + film + '/'
        thisfilmRep = requests.get(filmlink)
        thisfilmsoup = bs4.BeautifulSoup(thisfilmRep.text, 'html.parser')
        thisfilmthemes = thisfilmsoup.find_all(id='tab-genres')[0].find_all(class_='text-slug')
        themes_lst = []
        for a in thisfilmthemes:
            themes_lst.append(a.text)
        film_dict = {'film': film, 'themes': themes_lst}
        films_themes.append(film_dict)
    return films_themes
```
